```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('../data/raw/healthcare-dataset-stroke-data.csv')
```

```python
dataset.isnull().sum()
```

# Disbalance

```python
dataset['stroke'].value_counts(normalize=True)
```

# Correlation Analysis

```python
dataset_for_corr = dataset.select_dtypes(include=['float64', 'int64']).drop(columns=['id'])
plt.figure(figsize=(10, 8))
sns.heatmap(dataset_for_corr.corr().round(5), annot=True, cmap='coolwarm', fmt='.5f')
plt.title('Correlation Heatmap')
plt.show()
```

# Distributions of variables

```python
num_col = dataset.select_dtypes(include=['float64', 'int64']).columns.drop(["id", "stroke"])
for col in num_col:
    plt.figure(figsize=(10, 5))
    sns.histplot(data=dataset, x=col, hue='stroke', palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
cat_cols = dataset.select_dtypes(include=['object']).columns
for col in cat_cols:
    plt.figure(figsize=(10, 5))
    sns.countplot(data=dataset, x=col, hue='stroke', palette='Set2')
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```
